```python
#双向链表
class Node(object):
    def __init__(self,item):
        self.elem=item
        self.prev=None
        self.next=None
        
class DoubleLinkList(object):   #部分函数可以从SingleLinkList继承
    def __init__(self,node=None):
        self.__head=node
        
    def is_empty(self):
        return self.__head is None
    
    def length(self):
        #cur 游标，用来移动遍历节点
        cur=self.__head
        #count记录数量
        count=0
        while cur !=None:
            count+= 1
            cur=cur.next
        return count
    
    def travel(self):
        #遍历整个链表
        #cur 游标，用来移动遍历节点
        cur=self.__head
        while cur !=None:
            print(cur.elem,end=' ')
            cur=cur.next
        print()
            
    def add(self,item):
        #从头节点插入
        node= Node(item)
        node.next=self.__head
        self.__head=node
        node.next.prev=node
        
    
    def append(self,item):
        #从链表尾端插入
        node= Node(item)
        if self.is_empty():
            self.__head=node
        else:
            cur=self.__head
            while cur.next!=None:
                cur=cur.next
            cur.next=node
            node.prev=cur
        
        
    def insert(self,pos,item):
        #从任意位置插入
        #:param pos 从0开始
        node=Node(item)
        if pos<=0:
            self.add(item)
        elif pos>(self.length()-1):
            self.append(item)
        else:
            count=0
            cur=self.__head
            while count < pos:
                count+=1
                cur=cur.next
            node.next=cur
            node.prev=cur.prev
            cur.prev.next=node
            cur.prev=node
   
    
    def remove(self,item):
        cur=self.__head
        while cur!=None:
            if cur.elem==item:               
                if cur==self.__head:  #先判断此节点是否为头节点
                    self.__head=cur.next
                    if cur.next:
                        #判断链表是否只有一个节点
                        cur.next.prev=None
                    else:
                        self.__head=None
                elif cur.next==None: #判断是否为尾节点
                    cur.prev.next=None
                else:        
                    cur.prev.next=cur.next
                    cur.next.prev=cur.prev
                return
            else:
                cur=cur.next
            
           
    
    def search(self,item):
        cur=self.__head
        while cur.next!=None:
            if cur.elem==item:
                return True
            else:
                cur=cur.next
        return False
        
        
single_obj = DoubleLinkList()
print('single_obj.is_empty: ',single_obj.is_empty())
print('length:',single_obj.length())
single_obj.append(1)
print('single_obj.is_empty: ',single_obj.is_empty())
print('length:',single_obj.length())
single_obj.append(2)
single_obj.append(3)
single_obj.append(4)
print('single_obj.append 2 3 4,length:',single_obj.length())
single_obj.travel()
single_obj.add(10)
print('single_obj.add 10,length：',single_obj.length())
single_obj.travel()
single_obj.insert(3,11)
print('single_obj.insert(3,11),length:',single_obj.length())
single_obj.travel()
single_obj.search(5)
single_obj.remove(11)
print('single_obj.remove(11),length:',single_obj.length())
single_obj.travel()
single_obj.remove(6)
print('single_obj.remove(6),length:',single_obj.length())
single_obj.travel()
single_obj.remove(10)
print('single_obj.remove(10),length:',single_obj.length())
single_obj.travel() 
single_obj.remove(4)
print('single_obj.remove(4),length:',single_obj.length())
single_obj.travel()  
single_obj.remove(2)
print('single_obj.remove(2),length:',single_obj.length())
single_obj.travel() 
single_obj.remove(1)
print('single_obj.remove(1),length:',single_obj.length())
single_obj.travel() 
single_obj.remove(3)
print('single_obj.remove(3),length:',single_obj.length())
single_obj.travel() 
```
